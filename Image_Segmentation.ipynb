```python
import numpy as np
import cv2
```

```python
# To visualize images in Google Colab
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/OIP.jpg')
```

```python
img.shape
```

```python
img.dtype
```

```python
# Converting to float32
img_float = np.float32(img)
img_float.dtype
```

```python
# Reshaping the image
img_reshaped = img_float.reshape((-1, 3))
img_reshaped.shape
```

```python
num_clusters = 3 # number of clusters.
```

```python
# Criteria to stop the K-Means algorithm.
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
```

```python
# Number of times the algorithm is executed using different initial labellings.
attempts = 10
```

```python
# This flag is used to specify how initial centers are taken.
flag = cv2.KMEANS_RANDOM_CENTERS
```

```python
# Appplying K-Means algorithm
ret,label,center = cv2.kmeans(img_reshaped, num_clusters, None, criteria, attempts, flag)
```

```python
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((img.shape))
```

```python
print("Input Image")
cv2_imshow(img)

print("\nSegmented Image")
cv2_imshow(res2)
```
